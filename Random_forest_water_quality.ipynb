```python
df.info()
```

```python
df["ammonia"]=df["ammonia"].astype(float)
```

So we can't directly change the data type of the column ammonia because it contain object column as '#NUM!'

So we have to drop that all rows which contain these values

```python
df=df[df["ammonia"]!="#NUM!"]
```

```python
df['ammonia']=pd.to_numeric(df['ammonia'])
```

```python
df['is_safe']=pd.to_numeric(df['is_safe'])
```

```python
df["is_safe"].value_counts()
```

Visualizing the data set

```python
#Checking corretion for the data set
#with the help of the heat map


plt.figure(figsize=(10,8))
sns.heatmap(data=df.corr(),annot=False)
plt.title("Correlation between all the column in the data set with respect to target column")
plt.show()
```

For the above data we can clearly see that target column is unbaised

So to make it balanced we are going to import the resampling method

```python
from imblearn.over_sampling import RandomOverSampler
```

As it contain less count of the predicting value so it's  better to increase the predicting values for the target colummn. So we are using over_sampling method here

```python
sa=RandomOverSampler()
```

Importing train_split function

```python
from sklearn.model_selection import train_test_split
```

```python
df.columns
```

Deviding the data into 2 parts for inputs and outputs

```python
x=df[['aluminium', 'ammonia', 'arsenic', 'barium', 'cadmium', 'chloramine',
       'chromium', 'copper', 'flouride', 'bacteria', 'viruses', 'lead',
       'nitrates', 'nitrites', 'mercury', 'perchlorate', 'radium', 'selenium',
       'silver', 'uranium']]

y=df["is_safe"]
```

Resampling the input and output values

```python
x_resample,y_resample=sa.fit_resample(x,y)
```

Spliting data into traing and testing parts

```python
x_train,x_test,y_train,y_test=train_test_split(x_resample,y_resample,test_size=0.2,random_state=42)
```

```python
y_resample.value_counts()
```

RandomForestClassifier: An ensemble algorithm in scikit-learn that builds multiple decision trees and combines their results to improve accuracy and reduce overfitting.

accuracy_score & classification_report: Tools to evaluate model performance. accuracy_score gives the overall correctness, while classification_report shows precision, recall, and F1-score for each class.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,classification_report
```

```python
model=RandomForestClassifier(n_estimators=1,random_state=5)
```

```python
model.fit(x_train,y_train)
```

Predicting the y_pred value for x_test

```python
y_pred=model.predict(x_test)
y_train_pred=model.predict(x_train)
```

Calculating the accuracy

```python
print("Accuracy score for training data")
print(accuracy_score(y_train,y_train_pred))
print("=================================")
print("Accuracy score for testing data")
print(accuracy_score(y_test,y_pred))
```

Importing the joblib for to load and dump the model built

```python
import joblib

joblib.dump(model,"model.pkl")
```
